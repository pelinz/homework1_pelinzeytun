```python
import thinkstats2
import thinkplot
```

```python
valuable1=resp.fmarno
hist = thinkstats2.Hist(valuable1, label='marriages')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Age', ylabel='Pregnants')
```

```python
nevermarried = resp[resp.fmarno == 0]
married = resp[resp.fmarno >0]

nevermarried_hist = thinkstats2.Hist(nevermarried.totincr, label='nevermarried')
married_hist = thinkstats2.Hist(married.totincr, label='married')
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(nevermarried_hist, align='right', width=width)
thinkplot.Hist(married_hist, align='left', width=width)
thinkplot.Config(xlabel='weeks', ylabel='Count')
```
